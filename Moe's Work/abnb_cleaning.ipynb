```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
abnb = "../Resources/AB_NYC_2019.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
abnb_df = pd.read_csv(abnb, encoding="ISO-8859-1")
```

```python
# Show just the header
abnb_df.head()
```

```python
# Delete extraneous columns
abnb_df = abnb_df.drop(['host_name', 'number_of_reviews', 'last_review', 'reviews_per_month', 'calculated_host_listings_count'], axis = 1)
abnb_df.head(2)
```

```python
# Droping availabilities of 30days or less per year.
abnb_df.drop(abnb_df[abnb_df['availability_365'] < 30 ].index, inplace = True)
abnb_df.head()
```

```python
# Droping min_nights of over 365days.
abnb_df.drop(abnb_df[abnb_df['minimum_nights'] > 365 ].index, inplace = True)
abnb_df.head()
```

```python
abnb_df = abnb_df.sort_values("neighbourhood", ascending=False)
abnb_df.head()
```

```python
count = abnb_df['neighbourhood'].value_counts()
count
```

```python
# Grouping neigborhood count
grp_abnb_neigh = abnb_df.groupby(['neighbourhood'])
grp_abnb_neigh.count()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
abnb_df.to_csv("abnb_cleaned.csv", index=False, header=True)
```
